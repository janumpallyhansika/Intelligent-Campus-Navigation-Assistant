```python
import pandas as pd
import networkx as nx
import folium
import matplotlib.pyplot as plt
```

```python
campus_data = {
    "Location": [
        "Library",
        "CSE Block",
        "Admin Office",
        "Cafeteria",
        "Hostel",
        "Auditorium"
    ],

    "Latitude": [
        13.0827,
        13.0835,
        13.0840,
        13.0820,
        13.0830,
        13.0845
    ],

    "Longitude": [
        80.2707,
        80.2715,
        80.2720,
        80.2725,
        80.2730,
        80.2740
    ]
}

df = pd.DataFrame(campus_data)

df.to_csv(
    "campus_dataset.csv",
    index=False
)

print(df)
```

```python
G = nx.Graph()

connections = [

("Library","CSE Block",2),

("CSE Block","Admin Office",1),

("Library","Cafeteria",3),

("Cafeteria","Hostel",2),

("Hostel","Auditorium",1),

("Admin Office","Auditorium",2),

("CSE Block","Hostel",2)

]

for u,v,w in connections:

    G.add_edge(
        u,
        v,
        weight=w
    )

print("Graph Created")
```

```python
start = input(
    "Enter Start Location: "
)

end = input(
    "Enter Destination: "
)

path = nx.shortest_path(

    G,

    source=start,

    target=end,

    weight='weight'

)

distance = nx.shortest_path_length(

    G,

    source=start,

    target=end,

    weight='weight'

)

print()

print("Shortest Route:")

print(" -> ".join(path))

print()

print("Distance:", distance)
```

```python
m = folium.Map(

    location=[13.0830,80.2720],

    zoom_start=16
)

for i,row in df.iterrows():

    folium.Marker(

        [row["Latitude"],row["Longitude"]],

        popup=row["Location"]

    ).add_to(m)

route_points = []

for node in path:

    row = df[

        df["Location"] == node

    ]

    lat = float(

        row["Latitude"]

    )

    lon = float(

        row["Longitude"]

    )

    route_points.append(

        [lat,lon]

    )

folium.PolyLine(

    route_points,

    weight=6

).add_to(m)

m
```

```python
routes = []

weights = []

for i in range(

    len(path)-1

):

    u = path[i]

    v = path[i+1]

    routes.append(

        u+"-"+v

    )

    weights.append(

        G[u][v]['weight']

    )

plt.figure(

    figsize=(7,5)

)

plt.bar(

    routes,

    weights

)

plt.xlabel(

    "Route Segments"

)

plt.ylabel(

    "Distance"

)

plt.title(

    "Campus Route Analysis"

)

plt.show()
```

```python
plt.savefig(

    "route_graph.png"

)

m.save(

    "campus_navigation_map.html"

)

print(

    "Files Saved"

)
```
